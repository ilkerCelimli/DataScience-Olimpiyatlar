```python
def show_pie_chart(data):
    plt.figure()
    plt.pie(data["rating"].value_counts(), labels=data["rating"].value_counts().index, autopct="%1.1f%%",rotatelabels=True,textprops={"fontsize":8,"color":"white",},labeldistance=1.5)
    plt.legend(loc ="best")
    plt.show()
```

# Ratinge göre yönetmenler


```python
data["director"].value_counts()
```

```python
movie_list = data[data["director"] != "Unknown"]
movie_list.value_counts()
```

```python
directors_list = movie_list["director"].unique()
director_movie = []
temp = pd.get_dummies(movie_list, columns=["rating"])
for i in directors_list:
    #"G","NC-17","NR","NR","PG","PG-13","R","TV-14
    director_movie.append(
        {
            "director": i,
            "G": temp[temp["director"] == i]["rating_G"].sum(),
            "NR": temp[temp["director"] == i]["rating_NR"].sum(),
            "PG": temp[temp["director"] == i]["rating_PG"].sum(),
            "PG-13": temp[temp["director"] == i]["rating_PG-13"].sum(),
            "R": temp[temp["director"] == i]["rating_R"].sum(),
            "NC-17": temp[temp["director"] == i]["rating_NC-17"].sum(),
            "NC-14": temp[temp["director"] == i]["rating_TV-14"].sum(),}
    )
director_movie = pd.DataFrame(director_movie)
director_movie
```

# Çekilen filmlerin türlerine göre gruplama

```python
data.info()
data["type"].value_counts()
```

```python
listed_in = []
for i in data["listed_in"]:
    listed = i.split(",")
    for item in listed:
        item = item.strip()
        if item not in listed_in:
            listed_in.append(item)
listed_in
```

```python
listed_data = []
for i in listed_in:
    listed_data.append(
        {
            "type": i,
            "count": data["listed_in"].str.count(i).sum(),
        })
listed_data = pd.DataFrame(listed_data)
listed_data
```

```python
def show_listed_movie(func_data):
    plt.figure()
    plt.bar(func_data["type"], func_data["count"])
    plt.xlabel("Type")
    plt.xticks(rotation=90,ha="right")
    plt.ylabel("Count")
    plt.show()
```

```python
show_listed_movie(listed_data)
```

# Yönetmenlerin çektikleri filmilerin yılına göre gruplama

```python
movie_list = data[data["director"] != "Unknown"].value_counts()
movie_list
```

```python
result = (
    data.groupby(["release_year", "director"])
    .size()  # Yıl ve yönetmene göre film sayısını hesapla
    .reset_index(name="count")  # Sonuçları bir tabloya dönüştür
    .sort_values(by=["director", "release_year"], ascending=[True, False])  # Daha okunabilir hale getirmek için sıralama
)
result = result[result["director"] != "Unknown"]
result = result.groupby("release_year")["count"].sum().reset_index()
result
```

```python
plt.figure()

plt.bar(result["release_year"],result["count"],color="red")
plt.xlabel("Year")
plt.xticks(rotation=90,ha="right")
plt.show()
```
